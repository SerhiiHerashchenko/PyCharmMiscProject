```python
# --- ЗАДАНИЕ 3 ---
# Y = Y11 (Трив. життя чол.)
# X = X1, X3, X5, X9

y_col = 'Y11'
x_cols = ['X1', 'X3', 'X5', 'X9']
all_cols_task3 = [y_col] + x_cols

# 1. Корреляционная матрица и обратная к ней
corr_matrix = df[all_cols_task3].corr()
inv_corr = np.linalg.inv(corr_matrix.values)

n = len(df)
p = len(x_cols) # кол-во факторов (4)
k_fixed = p - 1 # кол-во фиксируемых переменных для частной корреляции (3)

print(f"--- Результаты Задания 3 (Y = {y_col}) ---")
print("-" * 95)
print(f"{'Фактор':<10} | {'r_part':<8} | {'t-stat':<8} | {'p-val':<8} | {'Дов. интервал (95%)':<20} | {'Вывод'}")
print("-" * 95)

# 2. Расчет частных коэффициентов и интервалов
for i, x_col in enumerate(x_cols):
    idx_in_matrix = i + 1

    # Элементы обратной матрицы
    c_0i = inv_corr[0, idx_in_matrix]
    c_00 = inv_corr[0, 0]
    c_ii = inv_corr[idx_in_matrix, idx_in_matrix]

    # Частная корреляция
    r_part = -c_0i / np.sqrt(c_00 * c_ii)

    # Проверка значимости (t-тест)
    dof = n - p - 1 # 52 - 4 - 1 = 47
    t_stat = r_part * np.sqrt(dof / (1 - r_part**2))
    p_val = 2 * (1 - stats.t.cdf(abs(t_stat), dof))

    # --- Расчет доверительного интервала (через Z-преобразование Фишера) ---
    z_val = 0.5 * np.log((1 + r_part) / (1 - r_part))
    se_z = 1 / np.sqrt(n - k_fixed - 3)
    u_crit = 1.96

    z_lower = z_val - u_crit * se_z
    z_upper = z_val + u_crit * se_z

    r_lower = (np.exp(2 * z_lower) - 1) / (np.exp(2 * z_lower) + 1)
    r_upper = (np.exp(2 * z_upper) - 1) / (np.exp(2 * z_upper) + 1)

    interval_str = f"{r_lower:.3f} ... {r_upper:.3f}"
    significance = "Значим" if p_val < 0.05 else "Не знач."
    print(f"{x_col:<10} | {r_part:<8.4f} | {t_stat:<8.4f} | {p_val:<8.4f} | {interval_str:<20} | {significance}")

# 3. Множественный коэффициент корреляции
R_mult = np.sqrt(1 - 1/inv_corr[0, 0])
R_sq = R_mult**2

# F-критерий Фишера
F_stat = (R_sq / p) / ((1 - R_sq) / (n - p - 1))
p_val_F = 1 - stats.f.cdf(F_stat, p, n - p - 1)

print("-" * 95)
print(f"Множественный R: {R_mult:.4f}")
print(f"Коэфф. детерминации R^2: {R_sq:.4f}")
print(f"F-статистика: {F_stat:.4f}, p-value: {p_val_F:.4e}")

sign_str = "ЗНАЧИМ" if p_val_F < 0.05 else "НЕ значим"
print(f"Вывод: Множественный коэффициент {sign_str}.")

# 4. Доверительный интервал для множественного R (Формула со страницы 16)
# Используется уровень значимости 2*alpha для 95% вероятности (1-2alpha) -> alpha = 0.025
alpha_conf = 0.025

# Степени свободы v1 и v2 (Стр. 16, формулы снизу)
# v1 = целая часть от ( (p + n*R^2)^2 / (p + 2*n*R^2) )
num_v1 = (p + n * R_sq)**2
den_v1 = p + 2 * n * R_sq
v1 = int(num_v1 / den_v1)
v2 = n - p - 1

# Квантили F-распределения
# F(alpha) - это верхняя критическая точка. В python ppf берет вероятность "слева".
# Значит F(alpha) из методички -> ppf(1 - alpha)
# F(1-alpha) из методички -> ppf(alpha)
F_val_upper_tail = stats.f.ppf(1 - alpha_conf, v1, v2) # F(alpha)
F_val_lower_tail = stats.f.ppf(alpha_conf, v1, v2)     # F(1-alpha)

# Формула интервала для R^2:
# Нижняя граница (делим на большее F)
numer_term = R_sq * (1 - (p + 1)/n)
lower_R2 = (numer_term / ((1 - R_sq) * F_val_upper_tail)) - (p/n)

# Верхняя граница (делим на меньшее F)
upper_R2 = (numer_term / ((1 - R_sq) * F_val_lower_tail)) - (p/n)



print(f"\nДоверительный интервал для R^2 (95%): {lower_R2:.4f} ... {upper_R2:.4f}")
print(f"Доверительный интервал для R  (95%): {np.sqrt(lower_R2):.4f} ... {np.sqrt(upper_R2):.4f}")
print("-" * 95)
```

```python
# --- ЗАДАНИЕ 4 ---
# Группа Y (U): Y11, Y12
# Группа X (V): X1, X3, X5, X9

cols_u = ['Y11', 'Y12']
cols_v = ['X1', 'X3', 'X5', 'X9']

def perform_cca(df_data, u_names, v_names, n_obs):
    cols_all = u_names + v_names
    corr_full = df_data[cols_all].corr().values

    p_dim = len(u_names)
    q_dim = len(v_names)

    R11 = corr_full[:p_dim, :p_dim]
    R12 = corr_full[:p_dim, p_dim:]
    R21 = corr_full[p_dim:, :p_dim]
    R22 = corr_full[p_dim:, p_dim:]

    R11_inv = np.linalg.inv(R11)
    R22_inv = np.linalg.inv(R22)

    M = R11_inv @ R12 @ R22_inv @ R21
    eigvals, eigvecs = np.linalg.eig(M)

    sorted_idx = np.argsort(eigvals)[::-1]
    eigvals = np.real(eigvals[sorted_idx])
    eigvecs = np.real(eigvecs[:, sorted_idx])

    can_corrs = np.sqrt(eigvals)

    # Векторы A
    A_final = []
    for i in range(len(can_corrs)):
        vec = eigvecs[:, i]
        scale = np.sqrt(vec.T @ R11 @ vec)
        A_final.append(vec / scale)
    A_final = np.array(A_final).T

    # Векторы B
    B_final = []
    for i in range(len(can_corrs)):
        r = can_corrs[i]
        vec_a = A_final[:, i]
        vec_b = (1/r) * R22_inv @ R21 @ vec_a
        B_final.append(vec_b)
    B_final = np.array(B_final).T

    return can_corrs, A_final, B_final

# 1. Запуск анализа
print("--- Задание 4: Канонический анализ ---")
can_r, A_vecs, B_vecs = perform_cca(df, cols_u, cols_v, n)

# Вывод коэффициентов и векторов
for i, r in enumerate(can_r):
    print("-" * 60)
    print(f"Каноническая пара #{i+1}")
    print(f"Коэффициент корреляции r = {r:.4f}")

    print("\n  Вектор A (веса Y):")
    for j, col_name in enumerate(cols_u):
        print(f"    {col_name}: {A_vecs[j, i]:.4f}")

    print("\n  Вектор B (веса X):")
    for j, col_name in enumerate(cols_v):
        print(f"    {col_name}: {B_vecs[j, i]:.4f}")

# 2. Оценка значимости (Критерий Бартлетта с критическими значениями)
print("-" * 60)
print("\n--- Оценка значимости канонических корреляций ---")
p_dim = len(cols_u)
q_dim = len(cols_v)
alpha = 0.05 # Уровень значимости

for i in range(p_dim):
    # Lambda статистика
    Lambda_stat = np.prod(1 - can_r[i:]**2)

    # Расчетное значение Хи-квадрат
    chi2_calc = -(n - 1 - 0.5*(p_dim + q_dim + 1)) * np.log(Lambda_stat)

    # Степени свободы
    df_chi = (p_dim - i) * (q_dim - i)

    # Критическое (табличное) значение Хи-квадрат
    # ppf(1-alpha) дает значение, слева от которого лежит (1-alpha) вероятность
    chi2_crit = stats.chi2.ppf(1 - alpha, df_chi)

    print(f"\nПроверка {i+1}-й канонической корреляции (r = {can_r[i]:.4f}):")
    print(f"  Хи-квадрат расчетное (Chi2_calc): {chi2_calc:.4f}")
    print(f"  Хи-квадрат критическое (Chi2_table) при df={df_chi}: {chi2_crit:.4f}")

    if chi2_calc > chi2_crit:
        print(f"  РЕЗУЛЬТАТ: {chi2_calc:.4f} > {chi2_crit:.4f} -> Корреляция ЗНАЧИМА (H0 отвергается)")
    else:
        print(f"  РЕЗУЛЬТАТ: {chi2_calc:.4f} < {chi2_crit:.4f} -> Корреляция НЕ значима (H0 принимается)")

# 3. Уравнения (для 1-й значимой пары)
print("-" * 60)
print("\n--- Уравнения первой пары канонических переменных ---")
print("U1 = ", end="")
terms_u = [f"{A_vecs[j,0]:.3f}*Z({cols_u[j]})" for j in range(len(cols_u))]
print(" + ".join(terms_u).replace("+ -", "- "))

print("V1 = ", end="")
terms_v = [f"{B_vecs[j,0]:.3f}*Z({cols_v[j]})" for j in range(len(cols_v))]
print(" + ".join(terms_v).replace("+ -", "- "))

# 4. Сокращение размерности
# --- ЧАСТЬ 2: ФУНКЦИЯ РАСЧЕТА CCA ---

def perform_cca(df_data, u_names, v_names):
    cols_all = u_names + v_names
    corr_full = df_data[cols_all].corr().values

    p_dim = len(u_names)

    R11 = corr_full[:p_dim, :p_dim]
    R12 = corr_full[:p_dim, p_dim:]
    R21 = corr_full[p_dim:, :p_dim]
    R22 = corr_full[p_dim:, p_dim:]

    try:
        R11_inv = np.linalg.inv(R11)
        R22_inv = np.linalg.inv(R22)
    except np.linalg.LinAlgError:
        return np.array([0]), np.array([]), np.array([])

    M = R11_inv @ R12 @ R22_inv @ R21
    eigvals, eigvecs = np.linalg.eig(M)

    sorted_idx = np.argsort(eigvals)[::-1]
    eigvals = np.real(eigvals[sorted_idx])
    eigvecs = np.real(eigvecs[:, sorted_idx])

    # Защита от отрицательных чисел под корнем из-за погрешности
    eigvals = np.maximum(eigvals, 0)
    can_corrs = np.sqrt(eigvals)

    # Векторы A
    A_final = []
    for i in range(len(can_corrs)):
        vec = eigvecs[:, i]
        denom = np.sqrt(vec.T @ R11 @ vec)
        if denom == 0: denom = 1e-10
        A_final.append(vec / denom)
    A_final = np.array(A_final).T

    # Векторы B
    B_final = []
    for i in range(len(can_corrs)):
        r = can_corrs[i]
        if r < 1e-6: # Защита от деления на 0
             vec_b = np.zeros(len(v_names))
        else:
            vec_a = A_final[:, i]
            vec_b = (1/r) * R22_inv @ R21 @ vec_a
        B_final.append(vec_b)
    B_final = np.array(B_final).T

    return can_corrs, A_final, B_final

# --- ЧАСТЬ 3: ИТЕРАТИВНОЕ СОКРАЩЕНИЕ РАЗМЕРНОСТИ ---

print("--- Итеративное сокращение размерности (Вариант 4) ---")
print(f"Исходный набор X: {cols_v}")

current_v_cols = cols_v.copy()
step = 0

while len(current_v_cols) > 1:
    step += 1
    print(f"\nШАГ {step}. Текущие переменные X: {current_v_cols}")

    # 1. Считаем корреляцию для текущего набора
    r_curr_list, _, B_curr = perform_cca(df, cols_u, current_v_cols)
    r_curr = r_curr_list[0]

    # 2. Ищем слабейшую переменную (мин вес в 1-м векторе)
    coeffs_v1 = B_curr[:, 0]
    min_idx = np.argmin(np.abs(coeffs_v1))
    candidate = current_v_cols[min_idx]

    print(f"  Макс. каноническая корреляция (текущая): {r_curr:.5f}")
    print(f"  Кандидат на удаление (мин. вес {coeffs_v1[min_idx]:.4f}): {candidate}")

    # 3. Считаем корреляцию БЕЗ кандидата
    test_v_cols = [c for c in current_v_cols if c != candidate]
    r_new_list, _, _ = perform_cca(df, cols_u, test_v_cols)
    r_new = r_new_list[0]
    print(f"  Макс. каноническая корреляция (без {candidate}): {r_new:.5f}")

    # 4. Тест значимости различия
    # z-преобразование Фишера
    z1 = 0.5 * np.log((1 + r_curr) / (1 - r_curr))
    z2 = 0.5 * np.log((1 + r_new) / (1 - r_new))

    # t-статистика
    t_calc = (z1 - z2) * np.sqrt((n - 3) / 2)
    # Критическое значение (двустороннее, alpha=0.05)
    t_crit = stats.t.ppf(1 - 0.05, n - 3)
    # В одностороннем тесте (мы проверяем только ухудшение) часто используют alpha,
    # но в примере использован стандартный подход.

    p_val_diff = (1 - stats.t.cdf(abs(t_calc), n - 3)) * 2 # Двусторонний P-value

    print(f"  Сравнение: t_calc = {t_calc:.4f}, t_crit = {t_crit:.4f} (p={p_val_diff:.4f})")

    if abs(t_calc) < t_crit:
        print(f"  РЕЗУЛЬТАТ: Различие НЕ значимо. Удаляем {candidate} и продолжаем.")
        current_v_cols = test_v_cols # Обновляем список переменных
    else:
        print(f"  РЕЗУЛЬТАТ: Различие ЗНАЧИМО. Удалять {candidate} нельзя.")
        print("  ОСТАНОВКА АЛГОРИТМА.")
        break

print("\n" + "="*50)
print(f"ИТОГОВЫЙ НАБОР ПЕРЕМЕННЫХ X: {current_v_cols}")
print("="*50)
```
