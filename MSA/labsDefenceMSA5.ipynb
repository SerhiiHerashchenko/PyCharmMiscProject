```python
import numpy as np

# Дані для Варіанта 4
# Y - Продуктивність праці (Y1)
Y = np.array([
    9.37, 13.17, 6.67, 5.68, 5.22, 10.02, 8.16, 8.78, 6.48, 10.44,
    7.65, 8.77, 7.00, 11.06, 9.02, 13.28, 9.27, 6.70, 6.69, 9.42,
    7.24, 5.39, 5.61, 5.59, 6.57, 6.54, 4.23, 5.22, 18.00, 11.03
])

# X - Коефіцієнт змінності устаткування (Feature 4)
# Значення взяті з відповідних комірок таблиці (діапазон 1.0-2.0)
X = np.array([
    1.4, 1.6, 1.4, 1.4, 1.0, 1.0, 1.6, 1.0, 1.7, 1.7,
    1.6, 1.7, 1.7, 1.6, 1.0, 1.6, 1.6, 1.4, 1.7, 1.3,
    1.7, 1.3, 1.0, 1.4, 1.3, 1.87, 1.47, 1.38, 1.41, 1.39
])

n = len(X)
Y_mean = np.mean(Y)
s = len(np.unique(X)) # Кількість груп

# Групування даних за унікальними значеннями X
groups = {}
for val in np.unique(X):
    groups[val] = Y[X == val]

# Кількість спостережень у групах та середні значення
v_j = [len(groups[val]) for val in groups]
y_mean_groups = [np.mean(groups[val]) for val in groups]

# Розрахунок загальної суми Y для перевірки
sum_y = np.sum(Y)

print(f"n (Загальна кількість спостережень) = {n}")
print(f"s (Кількість груп) = {s}")
print(f"Σy (Загальна сума Y) = {sum_y:.4f}")
print(f"ȳ (Загальне середнє Y) = {Y_mean:.4f}")
print(f"Групи X: {list(groups.keys())}")
print(f"v_j (Кількість у групі): {v_j}")
print(f"ȳ_j (Середнє у групі): {[f'{y:.4f}' for y in y_mean_groups]}")
print("-" * 50)

# 1. Розрахунок внутрішньогрупових сум квадратів відхилень
S_internal_groups = []
for idx, val in enumerate(groups.keys()):
    y_group = groups[val]
    y_mean_group = y_mean_groups[idx]
    S_internal = np.sum((y_group - y_mean_group)**2)
    S_internal_groups.append(S_internal)

# 2. Сума внутрішньогрупових сум квадратів
sum_S_internal = np.sum(S_internal_groups)

# 3. Внутрішньогрупова дисперсія (S²_ε)
S_epsilon_sq = sum_S_internal / n

print(f"Сума внутрішньогрупових сум квадратів (Σ S_внутр) = {sum_S_internal:.4f}")
print(f"S²_ε = (1/n) ⋅ Σ S_внутр = (1/30) ⋅ {sum_S_internal:.4f} ≈ {S_epsilon_sq:.4f}")

# 1. Розрахунок загальної суми квадратів
S_total_sum_sq = np.sum((Y - Y_mean)**2)

# 2. Загальна дисперсія (S²_y)
S_y_sq = S_total_sum_sq / n

print(f"S_заг (Загальна сума квадратів) = {S_total_sum_sq:.4f}")
print(f"S²_y = (1/n) ⋅ S_заг = (1/30) ⋅ {S_total_sum_sq:.4f} ≈ {S_y_sq:.4f}")

# Розрахунок коефіцієнта детермінації
K_d = 1 - (S_epsilon_sq / S_y_sq)

print(f"Отримаємо K_d(y;x) = 1 - S²_ε / S²_y")
print(f"K_d(y;x) = 1 - {S_epsilon_sq:.4f} / {S_y_sq:.4f} ≈ {K_d:.4f}")
```

```python
import numpy as np

# Використовуємо ті ж самі дані Варіанта 4
y = np.array([
    9.37, 13.17, 6.67, 5.68, 5.22, 10.02, 8.16, 8.78, 6.48, 10.44,
    7.65, 8.77, 7.00, 11.06, 9.02, 13.28, 9.27, 6.70, 6.69, 9.42,
    7.24, 5.39, 5.61, 5.59, 6.57, 6.54, 4.23, 5.22, 18.00, 11.03
])
x = np.array([
    1.4, 1.6, 1.4, 1.4, 1.0, 1.0, 1.6, 1.0, 1.7, 1.7,
    1.6, 1.7, 1.7, 1.6, 1.0, 1.6, 1.6, 1.4, 1.7, 1.3,
    1.7, 1.3, 1.0, 1.4, 1.3, 1.87, 1.47, 1.38, 1.41, 1.39
])
n = len(x)

# Розрахунок середніх значень
x_mean = np.mean(x)
y_mean = np.mean(y)

# Розрахунок сум (для перевірки)
x_sum = np.sum(x)
y_sum = np.sum(y)

print(f"n = {n}")
print(f"Сума x = {x_sum:.4f}")
print(f"x̄ = {x_mean:.4f}")
print(f"Сума y = {y_sum:.4f}")
print(f"ȳ = {y_mean:.4f}")

# Визначення знаків відхилень
x_minus_x_mean = np.sign(x - x_mean)
y_minus_y_mean = np.sign(y - y_mean)

# Підрахунок C (співпадаючих знаків) та H (розбіжних знаків)
C = np.sum((x_minus_x_mean * y_minus_y_mean) > 0)
H = np.sum((x_minus_x_mean * y_minus_y_mean) < 0)

print(f"C (Кількість співпадаючих знаків) = {C}")
print(f"H (Кількість розбіжних знаків) = {H}")

# Розрахунок коефіцієнта Фехнера
K_Fechner = (C - H) / n
print(f"Коефіцієнт Фехнера (K_Φ) = {K_Fechner:.4f}")
```

```python
import numpy as np
from scipy import stats

# Дані Варіанта 4
Y = np.array([
    9.37, 13.17, 6.67, 5.68, 5.22, 10.02, 8.16, 8.78, 6.48, 10.44,
    7.65, 8.77, 7.00, 11.06, 9.02, 13.28, 9.27, 6.70, 6.69, 9.42,
    7.24, 5.39, 5.61, 5.59, 6.57, 6.54, 4.23, 5.22, 18.00, 11.03
])
X = np.array([
    1.4, 1.6, 1.4, 1.4, 1.0, 1.0, 1.6, 1.0, 1.7, 1.7,
    1.6, 1.7, 1.7, 1.6, 1.0, 1.6, 1.6, 1.4, 1.7, 1.3,
    1.7, 1.3, 1.0, 1.4, 1.3, 1.87, 1.47, 1.38, 1.41, 1.39
])

n = len(X)
alpha = 0.1

# 1. Розрахунок необхідних сум
sum_x = np.sum(X)
sum_y = np.sum(Y)
sum_xy = np.sum(X * Y)
sum_x_sq = np.sum(X**2)
sum_y_sq = np.sum(Y**2)

# Чисельник
numerator = n * sum_xy - sum_x * sum_y
# Знаменник
denom_x_part = n * sum_x_sq - (sum_x)**2
denom_y_part = n * sum_y_sq - (sum_y)**2
denominator = np.sqrt(denom_x_part * denom_y_part)

# Коефіцієнт кореляції
r_hat = numerator / denominator

print(f"n = {n}")
print(f"Сума X (Σx) = {sum_x:.4f}")
print(f"Сума Y (Σy) = {sum_y:.4f}")
print(f"Сума XY (Σxy) = {sum_xy:.4f}")
print(f"Сума X² (Σx²) = {sum_x_sq:.4f}")
print(f"Сума Y² (Σy²) = {sum_y_sq:.4f}")
print("-" * 30)
print(f"Чисельник = {numerator:.4f}")
print(f"Знаменник = {denominator:.4f}")
print(f"r̂ = {r_hat:.4f}")

# Розрахунок t-статистики
r_sq = r_hat**2
t_stat = np.abs(r_hat) * np.sqrt((n - 2) / (1 - r_sq))

# Критичне значення t-критерію
df = n - 2
t_critical = stats.t.ppf(1 - alpha / 2, df)

print(f"r̂² = {r_sq:.4f}")
print(f"t_ст = {t_stat:.4f}")
print(f"t_крит при α=0.1: {t_critical:.4f}")

# Перетворення Фішера Z та інтервальні оцінки
Z_r = 0.5 * np.log((1 + r_hat) / (1 - r_hat))
S_Z = 1 / np.sqrt(n - 3)

z1 = Z_r - t_critical * S_Z
z2 = Z_r + t_critical * S_Z

r1 = (np.exp(2 * z1) - 1) / (np.exp(2 * z1) + 1)
r2 = (np.exp(2 * z2) - 1) / (np.exp(2 * z2) + 1)

print("-" * 30)
print(f"r1 (Нижня межа) ≈ {r1:.4f}")
print(f"r2 (Верхня межа) ≈ {r2:.4f}")
```

```python
import numpy as np
from scipy import stats

# Дані Варіанта 4
Y = np.array([
    9.37, 13.17, 6.67, 5.68, 5.22, 10.02, 8.16, 8.78, 6.48, 10.44,
    7.65, 8.77, 7.00, 11.06, 9.02, 13.28, 9.27, 6.70, 6.69, 9.42,
    7.24, 5.39, 5.61, 5.59, 6.57, 6.54, 4.23, 5.22, 18.00, 11.03
])
X = np.array([
    1.4, 1.6, 1.4, 1.4, 1.0, 1.0, 1.6, 1.0, 1.7, 1.7,
    1.6, 1.7, 1.7, 1.6, 1.0, 1.6, 1.6, 1.4, 1.7, 1.3,
    1.7, 1.3, 1.0, 1.4, 1.3, 1.87, 1.47, 1.38, 1.41, 1.39
])

n = len(X)
alpha = 0.1

# 1. Групування даних
unique_x, counts = np.unique(X, return_counts=True)
s = len(unique_x)

group_data = {}
for val in unique_x:
    group_data[val] = Y[X == val]

# 2. Розрахунок групових середніх (ȳ_j)
y_mean_groups = [np.mean(group_data[val]) for val in unique_x]
v_j = counts
Y_mean = np.mean(Y)

print(f"Кількість груп (s) = {s}")
print(f"Унікальні значення X (j) = {unique_x}")
print(f"Середні Y у групах = {[f'{y:.4f}' for y in y_mean_groups]}")
print("-" * 50)

# Розрахунок S_y_bar_sq (Міжгрупова дисперсія)
sum_v_y_bar_diff_sq = np.sum(v_j * (y_mean_groups - Y_mean)**2)
S_y_bar_sq = sum_v_y_bar_diff_sq / n

# Розрахунок S_y_sq (Загальна дисперсія)
S_y_sq = np.sum((Y - Y_mean)**2) / n

print(f"S̄²_y (Міжгрупова) = {S_y_bar_sq:.4f}")
print(f"S²_y (Загальна) = {S_y_sq:.4f}")

# Розрахунок кореляційного відношення
rho_tilde = np.sqrt(S_y_bar_sq / S_y_sq)
print(f"ρ̃_yx = {rho_tilde:.4f}")

# Розрахунок F-статистики
rho_tilde_sq = rho_tilde**2
df1 = s - 1
df2 = n - s
F_stat = (rho_tilde_sq / (1 - rho_tilde_sq)) * (df2 / df1)
F_critical = stats.f.ppf(1 - alpha, df1, df2)

print(f"F_ст = {F_stat:.4f}")
print(f"F_крит (α=0.1; {df1}; {df2}) = {F_critical:.4f}")

# Інтервальна оцінка (спрощена)
F_lower = stats.f.ppf(1 - alpha, n - s, s - 1)
F_upper = stats.f.ppf(alpha, n - s, s - 1)

lower_denom = n * (1 - rho_tilde_sq) * F_lower + (s - 1)
upper_denom = n * (1 - rho_tilde_sq) * F_upper + (s - 1)
rho_sq_lower = 1 / (n / lower_denom)
rho_sq_upper = 1 / (n / upper_denom)

print(f"ρ ∈ [{np.sqrt(rho_sq_lower):.4f}; {np.sqrt(rho_sq_upper):.4f}]")
```
